```python
df.drop(['DateTime','Holiday'],axis=1,inplace=True)
```

```python
df.isin(['?']).any()
```

```python
data_col=[col for col in df.columns if df[col].dtypes=='object']
data=df[data_col]
data.sample(10)
```

```python
for col in data.columns:
    data.drop(data.index[data[col]=='?'],inplace=True)
    
```

```python
data=data.apply(pd.to_numeric)
data=data.reset_index()
data.drop('index',axis=1,inplace=True)
```

```python
data.info()
```

```python
X = data.drop('SMPEP2',axis=1)
Y = data['SMPEP2']
```

Model building

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25)
```

```python
from sklearn.ensemble import RandomForestRegressor 

rf = RandomForestRegressor()
rf.fit(X_train,Y_train)
```

Model Evaluation

```python
from sklearn.metrics import mean_absolute_error
```

```python
train_preds=rf.predict(X_train)
test_preds=rf.predict(X_test)
print('Training MAE :',mean_absolute_error(Y_train,train_preds))
print('Test MAE : ',mean_absolute_error(Y_test,test_preds))
```
